```python
import pandas as pd
import numpy as np
from matplotlib import pylab

fulllen=False #
fulllen=True #


# run pep30-kplogo.ipynb first to get the input file
inputfile='Pep30.kpLogo.pwm.txt'
outputfile='Pep30-U-periodicity.pdf'

if fulllen:
    inputfile='Pep30.fulllen.kpLogo.pwm.txt'
    outputfile='Pep30-fulllen-U-periodicity.pdf'

pwm = np.loadtxt(inputfile)
a = np.transpose((pwm))
df = pd.DataFrame(a, columns = ['A','C','G','U'])

import logomaker

fig, ax = pylab.subplots()

fig.set_size_inches(9, 2, forward=True)

lg = logomaker.Logo(df,ax=ax)
# style using Logo methods
lg.style_spines(visible=False)
lg.style_spines(spines=['left', 'bottom'], visible=True)
#lg.style_xticks(rotation=90, fmt='%d', anchor=0)
#lg.ax.set_xticklabels('%d'%x for x in np.linspace(1,90,dtype=int,num=90))

ax.set_xticks(range(1,90,3))
ax.set_xticklabels('%d'%(x+1) for x in range(90))

#lg.style_xticks(anchor=1, spacing=3, rotation=0)


# style using Axes methods
ax.set_ylabel("Log10(P)", labelpad=-1)
ax.set_xlabel("\nCodon position", labelpad=-1)
ax.margins(0.2)

#lg.ax.xaxis.set_ticks_position('none')
#lg.ax.xaxis.set_tick_params(pad=-1)

for i in range(0,90,6):
    lg.highlight_position_range(pmin=i, pmax=i+2, color='whitesmoke')

fig.savefig(outputfile)
pylab.show()
```
