## Python script to analyze U-periodicity in the pep13 library

```python
# translate dna to protein

from Bio.Seq import Seq

# input count file
count_file='N39.counts.txt'

# output file
count_with_pep='Pep13.txt'

import os

if not os.path.exists(count_with_pep):
    outf = open(count_with_pep,'w')

    with open(count_file,'r') as f:
        n=0
        for line in f:
            n=n+1
            print(n,'\r',end='')
            flds = line.strip().split()
            # translate, continue after stop codon
            pep = Seq(flds[0]).translate()
            # get the first pep len
            pepLen = len(pep.split('*')[0])
            outf.write(flds[0]+'\t'+str(pep)+'\t'+str(pepLen)+'\t'+flds[1]+'\t'+flds[2]+'\t'+flds[3]+'\t'+flds[4]+'\n')
    outf.close()       
```

```python
# count frequency for each peptide length

letter2index={'A':0,'C':1,'G':2,'T':3}

import numpy as np

# dna seq length
L=39

# peptide len, 0 to 13
pL=int(L/3)+1

# for each peptide length, create 4 composition matrix: dna, rna, high, low

if not os.path.exists('Pep13freqs.npy'):
    freqs=[]
    for i in range(pL):
        freqs.append([])
        for j in range(4):
            freqs[i].append(np.zeros((4,L)))

    with open(count_with_pep,'r') as f:
        n=0
        for line in f:
            n=n+1
            print(n,'\r',end='')
            flds = line.strip().split()
            # AAAAAACGTCTTGAAATGAATGCGATAGTATGTTATGCTT        KKRLEMNAIVCYA   13      0       0       0       1
            #print(line)
            k = int(flds[2]) # peptide length
            for j in range(4):
                #if int(flds[j+1]) < 1000:
                #    continue
                for i in range(L):
                    freqs[k][j][letter2index[flds[0][i]],i] = freqs[k][j][letter2index[flds[0][i]],i] + int(flds[j+3])
                #print(j)
                #print(freqs[j])
            #break
    np.save("Pep13freqs", freqs)
else:
    freqs = np.load('Pep13freqs.npy')
```

```python
# normalize by total read counts to get frequency
total_read_count = [] # for each pep length and each sample
for i in range(pL):
    total_read_count.append([0]*4)

for k in range(pL): # for each peptide length
    for j in range(4): # for each sample
        # get total read count by looking at the first nucleotide
        total_read_count[k][j] = sum(freqs[k][j][:,0])
        # normalize each nucleotide position by total read count
        for i in range(39):
            freqs[k][j][:,i] = freqs[k][j][:,i]/total_read_count[k][j]  
np.save("Pep13freqs-normalized", freqs)
```

```python
import pandas as pd
import logomaker
from matplotlib import pylab
#from matplotlib import pylab
sample = 3
for peplen in range(pL):

    a = np.log2(np.transpose(((freqs[peplen][sample]+0.01)/(0.01+freqs[peplen][2]))))
    df = pd.DataFrame(a, columns = ['A','C','G','U'])

    fig, ax = pylab.subplots()

    fig.set_size_inches(8, 2, forward=True)

    lg = logomaker.Logo(df,ax=ax)
    # style using Logo methods
    lg.style_spines(visible=False)
    lg.style_spines(spines=['left', 'bottom'], visible=True)
    #lg.style_xticks(rotation=90, fmt='%d', anchor=0)
    #lg.ax.set_xticklabels('%d'%x for x in np.linspace(1,90,dtype=int,num=90))

    ax.set_xticks(range(1,39,3))
    ax.set_xticklabels('%d'%(x+1) for x in range(39))
    
    if peplen < 2:
        ax.set_ylim(-1.4,0.75)

    #lg.style_xticks(anchor=1, spacing=3, rotation=0)


    # style using Axes methods
    ax.set_ylabel("log2(Low/High GFP)", labelpad=-1)
    ax.set_xlabel("\nCodon position", labelpad=-1)
    ax.margins(0.2)

    #lg.ax.xaxis.set_ticks_position('none')
    #lg.ax.xaxis.set_tick_params(pad=-1)

    for i in range(0,39,6):
        lg.highlight_position_range(pmin=i, pmax=i+2, color='whitesmoke')

    fig.savefig('pep13-'+str(peplen)+'logo.pdf')
```
